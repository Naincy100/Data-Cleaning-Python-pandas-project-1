# concat join 2 or more dfs video 8

```python
import pandas as pd
india_weather = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
    "humidity": [80, 60, 78]
})
india_weather
```

```python
us_weather = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
    "humidity": [68, 65, 75]
})
us_weather
```

```python
df = pd.concat([india_weather,us_weather])
df
```

```python
df = pd.concat([india_weather,us_weather],ignore_index=True)
df
```

```python
df = pd.concat([india_weather,us_weather],keys=['india','us'])
df
```

```python
df.loc['india']
```

```python
df.loc['us']
```

```python
temperature_df = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
}, index=[0,1,2])
temperature_df
```

```python
windspeed_df = pd.DataFrame({
    "city": ["delhi","mumbai"],
    "windspeed": [7,12],
}, index=[1,0])
windspeed_df
```

```python
df = pd.concat([temperature_df,windspeed_df])
df
```

```python
df = pd.concat([temperature_df,windspeed_df],axis=1) # concat do not match anything, simply combines rows from both df rowise
df
```

```python
s = pd.Series(["humid",'rain','sunny'],name='event')
```

```python
df = pd.concat([temperature_df,s],axis=1)
df
```

# merge vieo 9

```python
import pandas as pd
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35]})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df2
```

```python
df3 = pd.merge(df1,df2,on='city') #like join (inner)
df3
```

```python
df4 = pd.DataFrame({
    "city": ["new york","chicago","orlando",'balimore'],
    "temperature": [21,14,35,22]})
df4
```

```python
df5 = pd.DataFrame({
    "city": ["chicago","new york","ndodoi"],
    "humidity": [65,68,35],
})
df5
```

```python
df6 = pd.merge(df4,df5,on='city')
df6
```

```python
df6 = pd.merge(df4,df5,on='city',how='outer')
df6
```

```python
df6 = pd.merge(df4,df5,on='city',how='left')
df6
```

```python
df6 = pd.merge(df4,df5,on='city',how='outer',indicator=True)
df6
```

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35,38],
    "humidity": [65,68,71, 75]
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "temperature": [21,14,35],
    "humidity": [65,68,71]
})
df2
```

```python
df3 = pd.merge(df1,df2,on='city')
df3
```

```python
df3 = pd.merge(df1,df2,on='city',suffixes = ('_left','_right'))
df3
```

# Pivot table 
It is used to summarize and aggregate data inside df

```python
df = pd.read_csv("weather2.csv")
df
```

```python
df3 = df.pivot_table(index='city',columns='date') #aggregates average values
df3
```

```python
df3 = df.pivot_table(index='city',columns='date',aggfunc = sum) #aggregates sum values
df3
```

```python
df3 = df.pivot_table(index='city',columns='date',aggfunc = "count") #aggregates count values
df3
```

```python
df3 = df.pivot_table(index='city',columns='date',margins=True) #aggregates values more avg it finds
df3
```
